```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
# Define the Lorenz system of differential equations
def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]




```

```python
# Set the parameters
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

```

```python
# Set the initial conditions
initial_conditions = [1.0, 0.0, 0.0]

```

```python
# Set the time span
t_span = (0, 100)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

```

```python
# Solve the system of equations
sol = solve_ivp(lorenz, t_span, initial_conditions, args=(sigma, rho, beta), t_eval=t_eval)

```

```python
# Plot individual time series for x, y, and z
plt.figure(figsize=(100, 5))
plt.subplot(131)
plt.plot(sol.t, sol.y[0], label='x')
plt.title('Lorenz Model - Time Series for x')
plt.xlabel('Time')
plt.ylabel('x')
plt.legend()
```

```python
# Plot individual time series for x, y, and z
plt.figure(figsize=(100, 5))
plt.subplot(132)
plt.plot(sol.t, sol.y[1], label='y', color='orange')
plt.title('Lorenz Model - Time Series for y')
plt.xlabel('Time')
plt.ylabel('y')
plt.legend()
```

```python
# Plot individual time series for x, y, and z
plt.figure(figsize=(100, 5))
plt.subplot(133)
plt.plot(sol.t, sol.y[2], label='z', color='green')
plt.title('Lorenz Model - Time Series for z')
plt.xlabel('Time')
plt.ylabel('z')
plt.legend()
```

```python
# Plot time series
plt.figure(figsize=(15, 5))
plt.plot(sol.t, sol.y[0], label='x')
plt.plot(sol.t, sol.y[1], label='y')
plt.plot(sol.t, sol.y[2], label='z')
plt.title('Lorenz Model - Time Series')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Plot 3D phase portrait
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2])
ax.set_title('Lorenz Model - 3D Phase Portrait')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

**What is Butterfly Effect**?

The **Butterfly Effect** is the concept that small changes or actions, even those seemingly insignificant, can lead to large and complex effects in a system over time.

1) Small actions or events can lead to significant and unpredictable outcomes.

2) Originates from chaos theory, suggesting tiny changes in initial conditions can cause large-scale effects.

3) Named after the idea that a butterfly's wing flap could set off a chain reaction resulting in a distant hurricane.

4) Highlights the sensitivity of complex systems to initial conditions.

5) Emphasizes the interconnectedness of events and their potential for widespread consequences.
